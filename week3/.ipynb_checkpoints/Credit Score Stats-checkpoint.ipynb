```python
import pandas as pd
```

```python
df = pd.read_csv('desafio1.csv')
```

```python
df.head()
```

```python
creditscore_stats = df.groupby('estado_residencia')['pontuacao_credito'].agg([pd.Series.mode, 'median','mean','std'])
creditscore_stats.columns = ['moda', 'mediana', 'media', 'desvio_padrao']
```

```python
creditscore_stats.to_json('submission.json', orient='index')
```
